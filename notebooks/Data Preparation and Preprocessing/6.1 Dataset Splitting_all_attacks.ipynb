```python
import glob
import pandas as pd
import numpy as np

# csv_files = sorted(glob.glob("../../data/*.csv"))[:3]  # use first 3 CSVs to keep size small
# print("Using files:", csv_files)


# # Reads each CSV file into a pandas DataFrame using list comprehension.
# # For each file 'f' in csv_files, 'pd.read_csv(f)' loads it into memory as a DataFrame.
# # 'dfs' will be a list containing 3 DataFrames.
# dfs = [pd.read_csv(f) for f in csv_files]


# # Concatenates all DataFrames in 'dfs' into one large DataFrame 'df'.
# # 'ignore_index=True' resets the row index so it runs sequentially from 0 to total_rows-1.
# df = pd.concat(dfs, ignore_index=True)

# print("\nAttack Types:")
# # Check unique attack types
# print(df['Label'].unique())

# # Prints the number of rows and columns in the combined DataFrame.
# # Example output: "Rows, Columns: (600000, 85)"
# print("Rows, Columns:", df.shape)

# print("\n\nCheckpoint 1")

# # Displays concise information about the DataFrame.
# # Output includes: 
# #   - Number of non-null values per column
# #   - Data types (int64, float64, object, etc.)
# #   - Memory usage
# df.info()            # shows dtypes and non-null counts

# print("\n\nCheckpoint 2")

# # Prints the first 50 column names to quickly inspect the dataset structure.
# # Example: Index(['Destination Port', 'Flow Duration', 'Total Fwd Packets', ...], dtype='object')
# # print(df.columns[:50])  # show first 50 column names

# # Configure pandas to display *all* columns (no truncation)
# # pd.set_option('display.max_columns', None)   # None = no column limit
# # pd.set_option('display.width', 2000)         # Expands line width for full display

# # Print all column names
# print("\nAll Columns Before Dropping the irrelevent columns:\n", df.columns.tolist())  # .tolist() shows them in a Python list format



##################################################################################################################




import glob
import pandas as pd
import numpy as np

# list all CSVs (sorted by filename)
csv_files = sorted(glob.glob("../../data/*.csv"))

# take benign from the first two files only
benign_source_files = csv_files[:2]
benign_filenames = [f.replace("\\", "/").split("/")[-1] for f in benign_source_files]

# define labels
attack_labels = ['Brute Force -XSS', 'SQL Injection', 'DoS attacks-SlowHTTPTest', 'Bot', 'DoS attacks-Hulk', 'Brute Force -Web']
benign_label_variants = ['Benign', 'Benign ']

attack_frames = []
benign_frames = []

chunk_size = 500_000  # adjust based on RAM

for f in csv_files:
    print("\nProcessing:", f)
    
    for chunk in pd.read_csv(f, chunksize=chunk_size):
        chunk['Label'] = chunk['Label'].str.strip()

        # attacks from all files
        attacks_here = chunk[chunk['Label'].isin(attack_labels)]
        if not attacks_here.empty:
            attack_frames.append(attacks_here)
        
        # benign only from the first two files
        if f.replace("\\", "/").split("/")[-1] in benign_filenames:
            benign_here = chunk[chunk['Label'].isin(benign_label_variants)]
            if not benign_here.empty:
                benign_frames.append(benign_here)

# combine results
attacks_all = pd.concat(attack_frames, ignore_index=True) if attack_frames else pd.DataFrame()
benign_twofiles = pd.concat(benign_frames, ignore_index=True) if benign_frames else pd.DataFrame()

df = pd.concat([attacks_all, benign_twofiles], ignore_index=True)

print("\nAttacks total:", attacks_all.shape)
print("Benign from two files:", benign_twofiles.shape)
print("Combined shape:", df.shape)
print("")
print(df['Label'].value_counts())

# df.to_csv("filtered_attacks_benign.csv", index=False)
# print("\nDone! Saved as filtered_attacks_benign.csv")
```

```python
# Correlation-based pruning
# getting rid of features in the dataset that have a strong relationship with each other, because they provide almost the same information to the machine learning model.


# Only keep numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=[np.number])

# Compute correlation matrix
corr = numeric_df.corr().abs()

# Identify highly correlated features
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]

# Drop correlated columns from original dataframe
df = df.drop(columns=to_drop)

print(f"Removed {len(to_drop)} highly correlated columns.")
```

```python
######### Handle missing values #########


# 1 Drop rows with too many missing values (>30%)
threshold = 0.30  # 30% missing allowed
missing_fraction_per_row = df.isnull().mean(axis=1)

# Count how many rows will be dropped
rows_to_drop = (missing_fraction_per_row > threshold).sum()
print(f"Dropping {rows_to_drop} rows with more than {threshold*100}% missing values...")

# Keep rows that have <=30% missing data
df = df.loc[missing_fraction_per_row <= threshold].reset_index(drop=True)


# 2 Replace numerical NaNs in the features with the median of the entire column
num_cols = df.select_dtypes(include=[np.number]).columns

for col in num_cols:
    if df[col].isnull().any():  # Only fill if NaNs exist
        median_value = df[col].median()  # You can change to df[col].mean() if desired
        df[col].fillna(median_value, inplace=True)

print("Missing values handled successfully!")
print("Remaining NaNs per column:")
print(df.isnull().sum()[df.isnull().sum() > 0])

```

```python
# Convert to numeric, coercing errors to NaN
df['Flow Byts/s'] = pd.to_numeric(df['Flow Byts/s'], errors='coerce')

# Fill NaNs with median
df['Flow Byts/s'].fillna(df['Flow Byts/s'].median(), inplace=True)

# Verify
print(df.isnull().sum())
```

```python
##### Handle infinite values #####

# 1 Identify numeric columns only
num_cols = df.select_dtypes(include=[np.number]).columns

# 2 Replace +inf and -inf with the column’s finite max/min values
for col in num_cols:
    finite_vals = df[col][np.isfinite(df[col])]
    if not finite_vals.empty:
        df[col].replace(np.inf, finite_vals.max(), inplace=True)
        df[col].replace(-np.inf, finite_vals.min(), inplace=True)

# 3 Fill NaNs (in numeric columns only) with each column’s median
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# 4 Verify remaining NaNs
print("Remaining NaNs per column (should be zero):")
print(df.isnull().sum())

# Sample of the first 3 rows
print(df.head(3))
```

```python
###### Handle Label Encoding #####

if 'Label' not in df.columns:
    raise SystemExit("No Label column found. Re-check CSVs.")

df['Label'] = df['Label'].apply(lambda x: 0 if x == 'Benign' else 1)
# print(df['target'].value_counts())
print(df['Label'].value_counts())

print(df.head(10))

```

```python
###### Feature Encoding ######

from sklearn.preprocessing import LabelEncoder

# Feature encoding is the process of converting categorical features (text or labels) into numeric form, which is required for most machine learning models.
# Most ML algorithms (like logistic regression, random forest, neural networks) cannot handle strings directly, so we need to convert them.

# Why We Use Feature Encoding?
# Machine learning models require numeric input.
# Helps the model understand categories without throwing errors.
# Label encoding is simple and works fine for features like Protocol that have a small number of distinct categories.
    
# Feature encoding is applied only to categorical input features, not the target (Label).
# After this step, the DataFrame is fully numeric and ready for ML.

# if c != 'Label' → excludes the Label column (because it’s already encoded).
cat_cols = [c for c in df.select_dtypes(include=['object','category']).columns if c != 'Label']
print("Categorical columns to encode:", cat_cols)

le = LabelEncoder()
for c in cat_cols:
    df[c] = le.fit_transform(df[c].astype(str))
    print("Encoded", c)

print(df.head())
```

```python
###### Feature Scaling ######

from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# 1. Separate features (X) and labels (y)
X = df.drop(columns=['Label'])
y = df['Label']

print("Number of numerical features before scaling:", X.shape[1])

# 2. Initialize Min-Max Scaler (scales features to range [0, 1])
scaler = MinMaxScaler()

# 3. Fit the scaler on X and transform X (returns a NumPy array)
X_scaled = scaler.fit_transform(X)

# 4. Convert scaled data back to a DataFrame to preserve column names
X = pd.DataFrame(X_scaled, columns=X.columns)

# 5. Quick sanity-check: show min and max for the first 10 columns
print("\nAfter scaling (first 10 columns min/max):")
print(X.describe().T[['min', 'max']].head(10))
```
